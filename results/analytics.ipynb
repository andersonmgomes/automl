```python
#analysing the results
DS_NAME = 'IRIS'

import os
root_path = os.path.abspath('') #current directory
print('Listing the result files list...')
results_file_list = [f for f in os.listdir(root_path) if ((f[-4:] == '.csv') and (f.find(DS_NAME)>-1))]
results_file_list
```

```python
#loading the dataframe list
import pandas as pd
result_list = [(file_name, pd.read_csv(file_name)) for file_name in results_file_list]
print(len(result_list), 'resuts loaded.')
```

```python
import numpy as np
main_metric_list = []
for result in result_list:
    main_metric_list.append(result[1].iloc[0,6])
print('média:', np.mean(main_metric_list))
print('max:', np.max(main_metric_list))
```
